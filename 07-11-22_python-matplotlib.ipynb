# Initiation à Matplotlib

permet de tracer des graphiques super convainquant

attention : ne pas mélanger les deux méthodes proposées par matplolib => créé des bugs !

```python
import numpy as np
```

```python
x = np.linspace(0, 2, 10)
y = x**2
print(x)
print(y)
```

## pyplot (fonction)

```python
import matplotlib.pyplot as plt
```

attention : x et y doivent avoir le même nombre de points => renvoie une erreur dans le cas contraire

```python
plt.plot(x, y)
#plt.show() si on ne travaille pas dans notebook jupiter
```

```python
plt.scatter(x, y) #nuage de points
```

### customiser une courbe

```python
plt.plot(x, y, c='red', lw=3, ls='--')
# c = couleur de la courbe
# lw = épaisseur de la courbe
# ls = aspect de la courbe
```

### Cycle de vie d'une figure

```python
plt.figure(figsize=(12, 8)) # créé une page blanche qui permet de modifier la taille du graphique
plt.plot(x , y, label='quadratique') # label donne un nom à la courbe
plt.plot(x, x**3, label='cubique')
plt.title('figure 1') #donne un titre au graphique
plt.xlabel('axe X') #titre à l'axe des X
plt.ylabel('axe Y') #titre à l'axe des Y
plt.legend() #légende
plt.savefig('cycle de vie') #enregistre le fichier sur l'ordi
```

## Créer plusieurs graphiques

```python
plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x , y , c='red')
plt.title('graph 1')
plt.subplot(2,2,2)
plt.plot(x, y, c='yellow')
plt.subplot(2, 2, 3)
plt.plot(x , y , c='blue')
plt.subplot(2,2,4)
plt.plot(x, y, c='green')
plt.title('grap 4')
```

## OOP (orienté objet)

déconseillé => 90% peut être réalisé avec pyplot <br/>
mais offre plus de possibilité

```python
fig, ax = plt.subplots() #créer un tableau
ax.plot(x, y)
```

```python
fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(x, y, c='red')
ax[1].plot(x, np.sin(x), c='yellow')
```

### exercice

```python
dataset = {f"experience{i}": np.random.randn(100) for i in range(4)}
dataset['experience0']
```

```python
def graphique(data):
    plt.figure(figsize=(12, 8))
    n = len(data)
    for i in data:
        plt.subplot(n, 1, int(i[-1]) +1)
        plt.plot([j for j in range(len(data[i]))], data[i])
        plt.title(i) 
graphique(dataset)
```

```python
#correction
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12,8))
    for k, i in zip(data.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(data[k])
        plt.title(k)
graphique(dataset)
```
