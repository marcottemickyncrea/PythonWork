# Initiation à Matplotlib

permet de tracer des graphiques super convainquant

attention : ne pas mélanger les deux méthodes proposées par matplolib => créé des bugs !

```python
import numpy as np
```

```python
x = np.linspace(0, 2, 10)
y = x**2
print(x)
print(y)
```

## pyplot (fonction)

```python
import matplotlib.pyplot as plt
```

attention : x et y doivent avoir le même nombre de points => renvoie une erreur dans le cas contraire

```python
plt.plot(x, y)
#plt.show() si on ne travaille pas dans notebook jupiter
```

```python
plt.scatter(x, y) #nuage de points
```

## customiser une courbe

```python
plt.plot(x, y, c='red', lw=3, ls='--')
# c = couleur de la courbe
# lw = épaisseur de la courbe
# ls = aspect de la courbe
```

## Cycle de vie d'une figure

```python
plt.figure(figsize=(12, 8)) # créé une page blanche qui permet de modifier la taille du graphique
plt.plot(x , y, label='quadratique') # label donne un nom à la courbe
plt.plot(x, x**3, label='cubique')
plt.title('figure 1') #donne un titre au graphique
plt.xlabel('axe X') #titre à l'axe des X
plt.ylabel('axe Y') #titre à l'axe des Y
plt.legend() #légende
plt.savefig('cycle de vie') #enregistre le fichier sur l'ordi
```
