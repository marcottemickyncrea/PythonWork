## cross validation (plusieurs m√©thodes)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

X= iris.data
y= iris.target

print(X.shape)
plt.scatter(X[:, 0], X[:, 1], c=y, alpha = 0.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import StratifiedKFold #choix le plus sur
from sklearn.model_selection import GroupKFold
```

```python
cv = KFold(5, random_state=0, shuffle=True) #d√©coupe/split en 5 split, les tas se suivent => tr√®s utile pour les r√©gressions
cross_val_score(KNeighborsClassifier(), X, y , cv=cv).mean()
```

```python
cv = LeaveOneOut() #chque √©l√©ment sert √† √©valuer => consomme beaucoup de puissance de l'ordinateur !
cross_val_score(KNeighborsClassifier(), X, y , cv = cv).mean()
```

```python
cv = ShuffleSplit(4, test_size=0.2) # recoupe et m√©lange plusieurs fois
cross_val_score(KNeighborsClassifier(), X, y , cv =cv).mean()
```

```python
cv = StratifiedKFold() # ordi r√©partit les class pour fait des tas proportionn√© => √† pr√©f√©rer
cross_val_score(KNeighborsClassifier(), X, y , cv =cv).mean()
```

```python
cv = GroupKFold(5).get_n_splits(X, y, groups=X[:, 0])
cross_val_score(KNeighborsClassifier(), X, y , cv =cv).mean()
```
